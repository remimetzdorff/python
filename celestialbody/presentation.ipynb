```python
# imports classiques

import matplotlib.pyplot as plt
import numpy as np
import datetime # Pour les dates
date = datetime.datetime(2021, 2, 26, 0, 0)

# Pour l'affichage des graphiques dans le notebook 
%matplotlib inline
```

```python
# LA classe

from celestialbody import CelestialBody
```

```python
# création d'un objet celeste : la Terre
# les paramètres sont récupérés d'après les données disponibles ici https://ssd.jpl.nasa.gov/

earth = CelestialBody("EM Bary")
```

```python
# Choix de la date pour le calcul de position

earth.date = date
```

```python
# Coordonnées de la position de la Terre dans le référentiel ecliptique héliocentrique

earth.position
```

```python
earth.orbit
```

```python
# Représentation de l'orbite de la Terre
x,y,z = earth.position
X,Y,Z = earth.orbit

# Préparation de la figure
plt.figure(figsize=(12,8))
sps = (1,2)
ax1 = plt.subplot2grid(sps, (0,0))
ax2 = plt.subplot2grid(sps, (0,1), projection='3d')

# Représentation de la Terre et du Soleil
ax1.scatter(0,0, marker="o", color="gold", label="Soleil")
ax1.scatter(x,y, marker="o", color="C2", label="Terre")
ax1.plot(X,Y, "C2", linewidth=.5)
ax1.annotate("Terre", (x,y), color="C2", textcoords="offset points", xytext=(2,2),
             fontsize="medium", horizontalalignment='left', verticalalignment='bottom', alpha=1)

ax2.scatter(0,0,0, marker="o", color="gold", label="Soleil")
ax2.scatter(x,y,z, marker="o", color="C2", label="Terre")
ax2.plot(X,Y,Z, "C2", linewidth=.5)

# Gestion des limites, etc.
a_lim = 1.5
ax1.set_xlim(-a_lim,a_lim)
ax1.set_ylim(-a_lim,a_lim)
ax1.set_aspect("equal")

ax2.set_xlim(-a_lim,a_lim)
ax2.set_ylim(-a_lim,a_lim)
ax2.set_zlim(-a_lim,a_lim)

ax2.legend()
```

```python
# Au tour des huit planètes du système solaire

mercury = CelestialBody("Mercury")
venus   = CelestialBody("Venus")
earth   = CelestialBody("EM Bary")
mars    = CelestialBody("Mars")
jupiter = CelestialBody("Jupiter")
saturn  = CelestialBody("Saturn")
uranus  = CelestialBody("Uranus")
neptune = CelestialBody("Neptune")

planets = [mercury, venus, earth, mars, jupiter, saturn, uranus, neptune]

a_lim = 2

# Préparation de la figure
plt.figure(figsize=(12,8))
sps = (1,2)
ax1 = plt.subplot2grid(sps, (0,0))
ax2 = plt.subplot2grid(sps, (0,1), projection='3d')

ax1.scatter(0,0, marker="o", color="gold", label="Soleil")
ax2.scatter(0,0,0, marker="o", color="gold", label="Soleil")

for i, planet in enumerate(planets):
    planet.date = date
    x,y,z = planet.position
    X,Y,Z = planet.orbit

    # Représentation de la Terre et du Soleil
    ax1.scatter(x,y, marker="o", color="C"+str(i), label=planet.name)
    ax1.plot(X,Y, "C"+str(i), linewidth=.5)
    if planet.a < a_lim:
        ax1.annotate(planet.name, (x,y), color="C"+str(i), textcoords="offset points", xytext=(2,2),
                     fontsize="medium", horizontalalignment='left', verticalalignment='bottom', alpha=1)

    ax2.scatter(x,y,z, marker="o", color="C"+str(i), label=planet.name)
    ax2.plot(X,Y,Z, "C"+str(i), linewidth=.5)

# Gestion des limites, etc.
ax1.set_xlim(-a_lim,a_lim)
ax1.set_ylim(-a_lim,a_lim)
ax1.set_aspect("equal")

ax2.set_xlim(-a_lim,a_lim)
ax2.set_ylim(-a_lim,a_lim)
ax2.set_zlim(-a_lim,a_lim)

ax2.legend()
```

```python
earth = CelestialBody("EM Bary")
earth.date = date
x,y,z = earth.position
X,Y,Z = earth.orbit

ceres = CelestialBody("Ceres")
ceres.date = date
x1,y1,z1 = ceres.position
X1,Y1,Z1 = ceres.orbit

mars = CelestialBody("Mars")
mars.date = date
x2,y2,z2 = mars.position

# Préparation de la figure
plt.figure(figsize=(12,8))
sps = (1,2)
ax1 = plt.subplot2grid(sps, (0,0))
ax2 = plt.subplot2grid(sps, (0,1), projection='3d')

# Représentation de la Terre et du Soleil
ax1.scatter(0,0, marker="o", color="gold", label="Soleil")
ax1.scatter(x,y, marker="o", color="C2", label="Terre")
ax1.plot(X,Y, "C2", linewidth=.5)
ax1.annotate("Terre", (x,y), color="C2", textcoords="offset points", xytext=(2,2),
             fontsize="medium", horizontalalignment='left', verticalalignment='bottom', alpha=1)
ax1.scatter(x1,y1, marker="o", color="C3", label="Ceres")
ax1.plot(X1,Y1, "C3", linewidth=.5)
ax1.annotate("Ceres", (x1,y1), color="C3", textcoords="offset points", xytext=(2,2),
             fontsize="medium", horizontalalignment='left', verticalalignment='bottom', alpha=1)

ax1.scatter(x2,y2, marker="o", color="C1", label="Mars")

ax2.scatter(0,0,0, marker="o", color="gold", label="Soleil")
ax2.scatter(x,y,z, marker="o", color="C2", label="Terre")
ax2.plot(X,Y,Z, "C2", linewidth=.5)

# Gestion des limites, etc.
a_lim = 4
ax1.set_xlim(-a_lim,a_lim)
ax1.set_ylim(-a_lim,a_lim)
ax1.set_aspect("equal")

ax2.set_xlim(-a_lim,a_lim)
ax2.set_ylim(-a_lim,a_lim)
ax2.set_zlim(-a_lim,a_lim)

ax2.legend()
```
